```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
ora = pickle.load(open('oracle.pkl','r'))
```

```python
train_regions_list = [['Austin'],['Boulder'],['SanDiego'],
['Austin','SanDiego'],['Austin','Boulder'],['Boulder','SanDiego'],['Austin','Boulder','SanDiego']]
```

```python
train_order = ['_'.join(x) for x in train_regions_list]
```

```python
train_order
```

```python
pd.DataFrame(ora['hvac'])[train_order].plot(kind='bar',rot=0)
plt.ylabel("Error \n(Lower is better)")
plt.xlabel("Test region")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=2)
plt.title('Oracle performance for HVAC')
```

```python
pd.DataFrame(ora['fridge'])[train_order].plot(kind='bar',rot=0)
plt.ylabel("Error \n(Lower is better)")
plt.xlabel("Test region")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=2)
```
