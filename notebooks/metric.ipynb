```python
def scale_0_1(ser, minimum=None, maximum=None):
    if minimum is not None:
        pass
    else:
        minimum = ser.min()
        maximum = ser.max()
    return (ser - minimum).div(maximum - minimum)

def normalise(df):
    new_df = df.copy()
    max_aggregate = df[["aggregate_%d" % i for i in range(1, 13)]].max().max()
    min_aggregate = df[["aggregate_%d" % i for i in range(1, 13)]].min().min()
    new_df[["aggregate_%d" % i for i in range(1, 13)]] = scale_0_1(df[["aggregate_%d" % i for i in range(1, 13)]],
                                                                   min_aggregate, max_aggregate)
    for col in ['area', 'total_occupants', 'num_rooms', 'ratio_min_max',
                'skew', 'kurtosis', 'variance', 'ratio_difference_min_max', 'p_25',
                'p_50', 'p_75']:
        new_df[col] = scale_0_1(df[col])
    return new_df

df = normalise(df)
```

```python
features = flatten([['aggregate_%d' %i for i in range(1, 13)], ['area','num_rooms','total_occupants']])
```

```python
appliance="fridge"
if appliance=="hvac":
    start, stop=5, 11
else:
    start, stop=1, 13
appliance_df = df.ix[df[['%s_%d' %(appliance,month) for month in range(start,stop)]].dropna().index]
```

```python
k=5
out = {}
for home in appliance_df.index:
    distance = np.square(df[features]-appliance_df.ix[home][features]).sum(axis=1)
    distance.sort_values(inplace=True)
    distance = distance.drop([home])
    nghbrs = distance.head(k).index
    pred = appliance_df[['%s_%d' %(appliance,month) for month in range(start,stop)]].ix[nghbrs].mean()
    out[home] = pred
```

```python
pred_df = pd.DataFrame(out).T
```
