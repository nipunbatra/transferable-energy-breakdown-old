```python
def rename_column(df):
    for i in range(1, 13):
        df = df.rename(columns = {'aggregate_' + str(i): 'Aggregate_'+ str(i),
                        'hvac_' + str(i): 'HVAC_' + str(i),
                        'fridge_' + str(i): 'Fridge_' + str(i),
                        'mw_' + str(i):'Micro-wave_' + str(i),
                        'dw_' + str(i):'Dish Washer_' + str(i),
                        'wm_' + str(i): 'Washing Machine_' + str(i),
                        'oven_' + str(i): 'Oven_' + str(i)})
#     print df
    return df

Appliance_order = ['Aggregate', 'HVAC', 'Fridge', 'Micro-wave', 'Dish Washer', 'Washing Machine', 'Oven']
```

```python
source_df, source_dfc, source_tensor, source_static = create_region_df_dfc_static(source, year)
source_df = rename_column(source_df)
source_dfc = rename_column(source_dfc)
```

```python
# latexify(columns=1)
year = 2014
means = {}
for source in ['Austin','SanDiego']:
    source_df, source_dfc, source_tensor, source_static = create_region_df_dfc_static(source, year)
    source_df = rename_column(source_df)
    source_dfc = rename_column(source_dfc)
    means[source] = {}
    for appliance in Appliance_order:
        means[source][appliance] = source_df[['{}_{}'.format(appliance, month) for month in range(1, 13)]].mean()
        means[source][appliance].index = range(1, 13)

fig, ax  = plt.subplots(ncols=2)
for col, region in enumerate(['Austin','SanDiego']):
    (pd.DataFrame(means[region])[Appliance_order[1:]]).plot(kind='area', stacked=True, ax=ax[col], legend=False)
    pd.DataFrame(means[region])['Aggregate'].plot(ax=ax[col], title="{}".format(region), legend=False)
ax[0].set_ylim((0, 1400))    
ax[1].set_ylim((0, 1400)) 
#ax[2].set_ylim((0, 800))    
format_axes(ax[0])
format_axes(ax[1])
l = plt.legend(bbox_to_anchor=(1.1, 1))

t = fig.text(0.5, -0.02, "Month of year")

q = ax[0].set_ylabel("Energy consumption (kWh)")
# plt.savefig("../figures/region-energy.pdf", bbox_inches="tight", bbox_extra_artists=[l, t, q])
```
