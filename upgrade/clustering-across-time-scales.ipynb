```python
temp = pd.DataFrame(index=d.index)
```

```python
pd.value_counts(temp.index.month)[1]
```

```python
baseline = {}
for home in d.T.index:
    temp_home = [] 
    for month in range(1, 13):
        q = pd.value_counts(temp.index.month)[month]
        temp_home_month = np.zeros(shape=q)
        temp_home_month[:] = dm[home].values[month-1]/(q)
        temp_home.append(temp_home_month)
    temp_home = np.hstack(np.array(temp_home))
    baseline[home] = pd.Series(temp_home, index=d.index)
```

```python
dm[home].values[month-1]/(pd.value_counts(temp.index.month)[1])
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nbrs = NearestNeighbors(n_neighbors=3, algorithm='ball_tree').fit(dm.T)
distances, indices = nbrs.kneighbors(dm.T)
```

```python
pred = {}
ngh = {}
for i, home in enumerate(dm.T.index[:4]):
    pred[home] = d[dm.T.index[indices[:,1:]][i]].mean(axis=1)
    ngh[home] = dm.T.index[indices[:,1:]][i]
```

```python
pred_df = pd.DataFrame(pred)
```

```python
import matplotlib.gridspec as gridspec
def plot_home(pred_df, home, indices, d):
    fig = plt.figure(figsize=(16, 4))
    gs = gridspec.GridSpec(1, 2,
                       width_ratios=[3, 1])
                      
    ax = [plt.subplot(gs[0]), plt.subplot(gs[1])]

    f = d[ngh[home]]
    #f.plot(alpha=0.8, lw=0.5, color='k',legend=False, ax=ax[0])
    edf = pd.DataFrame({"GT":d[home], 
                        "Pred":pred[home]
                       ,"Baseline":baseline[home]})
    edf.plot(ax=ax[0])
    
    dm[home].plot(ax=ax[1])
    dm[ngh[home]].plot(ax=ax[1], color='k', legend=True)
    ax[0].set_title("Daily energy prediction")
    ax[1].set_title("Nearest neighbors (black) on \nmonthly aggregate energy")
    plt.suptitle("Home number {}".format(home))
```

```python
plot_home(pred_df, 22, indices, d)
```
